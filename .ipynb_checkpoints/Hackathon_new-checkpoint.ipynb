```python
insurance.columns

# age - The age of the primary beneficiary (insured individual)

# sex - The gender of the policyholder (male or female)

# bmi - Body Mass Index. An indicator of a person's weight relative to their height,
# representing objectively high or low body weight. The ideal range is 18.5–24.9

# children - The number of children covered by the health insurance (number of dependents)

# smoker - Smoker (yes/no). Indicates whether the beneficiary is a smoker

# region - The region of residence of the beneficiary in the US (e.g., southwest, southeast, northwest, northeast)

# charges - Individual medical costs billed by health insurance plans (the target variable often predicted in analysis)
```

```python
# transform 'sex' and 'smokers' from categorical features to numerical features to use statistical analysis

# transform 'sex'
sex_mapping = {'female': 0, 'male': 1}
insurance['sex_encoded'] = insurance['sex'].map(sex_mapping)

# transform 'smoker'
smoker_mapping = {'no': 0, 'yes': 1}
insurance['smoker_encoded'] = insurance['smoker'].map(smoker_mapping)
```

```python
# transform 'region' feature from nominal to numeric with One-Hot Encoding

region_dummies = pd.get_dummies(insurance['region'], prefix='region')

insurance = pd.concat([insurance, region_dummies], axis=1) # merge with 'insurance' dataset
# if we will make 'bonus', we can transform booleans to numerical features .astype(int)
```

```python
# standartization for 'age' and 'bmi' with minmax, because this feature has wide range of values

insurance['age_stand'] = (insurance['age'] - insurance['age'].mean()) / insurance['age'].std()

insurance['bmi_stand'] = (insurance['bmi'] - insurance['bmi'].mean()) / insurance['bmi'].std()
```

```python
# normalization for 'charges' with minmax, because this feature has wide range of values

insurance['charges_norm'] = (insurance['charges'] - insurance['charges'].min()) / (insurance['charges'].max() - insurance['charges'].min())
```

```python
# Summary statistics for charges by smoker status
smoker_charges = insurance.groupby('smoker')['charges'].describe()
smoker_charges['median'] = insurance.groupby('smoker')['charges'].median()
smoker_charges = smoker_charges.round(2)
smoker_charges

```

```python
# Summary statistics for charges by BMI
bmi_charges = insurance.groupby('bmi')['charges'].describe()
bmi_charges['median'] = insurance.groupby('bmi')['charges'].median()
bmi_charges = bmi_charges.round(2)
bmi_charges
```

```python
# Summary statistics for charges by children
children_charges = insurance.groupby('children')['charges'].describe()
children_charges['median'] = insurance.groupby('children')['charges'].median()
children_charges = children_charges.round(2)
children_charges
```

# Statistical Analysis And Data Visualization

---
## Our hypotheses:

- H1. Smokers have higher charges
- H2. People with BMI higher then normal have higher charges
- H3. People with 3 and more children have higher charges
---

```python
# preparation for correlation matrix (heatmap)
# this will give us a visual representation of the correlations between all the numerical variables in the dataset.
insurance_copy = insurance.copy()
columns_to_drop = ['age', 'sex', 'bmi', 'smoker', 'region', 'charges']
insurance_copy.drop(columns=columns_to_drop, inplace=True)
```

```python
# plot a correalation matrix
corr_matrix = insurance_copy.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# correlation for smokers

smoker_corr = spearmanr(insurance['smoker_encoded'], insurance['charges'])

# correlation for bmi

bmi_corr = spearmanr(insurance['bmi_stand'], insurance['charges'])

# correlation for bmi

ch_corr = spearmanr(insurance['children'], insurance['charges'])

print(f'''Spearman correlation results:
For Smokers : {smoker_corr}
For BMI : {bmi_corr}
For Number of Children : {ch_corr}''')
```

---
**Conclusions about Spearman correlation**:
- Smoker - This is the most important factor. The relationship between smoking status and spending is very strong and highly statistically significant. Smokers have significantly higher spending ranks than non-smokers.
- BMI level - Although BMI and expenditure ranks increase together, this increase is not dramatic. The very low P-value suggests that this weak association is not random and is present in the general population.
- Number og children - As the number of children increases, the ranks of expenditures increase slightly. The low p-value confirms that this weak relationship is statistically significant.
---

```python
# count mean for Smokers for barplot
mean_non_smoker = insurance[insurance['smoker_encoded'] == 0]['charges'].mean()
mean_smoker = insurance[insurance['smoker_encoded'] == 1]['charges'].mean()
groups = ['Non-smoker', 'Smoker']
means = [mean_non_smoker, mean_smoker]

# create groups for bmi for boxplot
charges_lower_bmi = insurance[insurance['bmi'] < 18.5]['charges']
charges_normal_bmi = insurance[(insurance['bmi'] >= 18.5) & (insurance['bmi'] <= 24.9)]['charges']
charges_higher_bmi = insurance[(insurance['bmi'] >= 25) & (insurance['bmi'] <= 29.9)]['charges']
charges_obesity_bmi = insurance[insurance['bmi'] >= 30]['charges']

# create groups for children for boxplot
charges_no_ch = insurance[insurance['children'] == 0]['charges']
charges_onetwo_ch = insurance[(insurance['children'] == 1) | (insurance['children'] == 2)]['charges']
charges_threemore_ch = insurance[(insurance['children'] == 3) | (insurance['children'] == 4) | (insurance['children'] == 5)]['charges']

# Median setting for bmi
median_style = {
    'color': 'royalblue',   # color of the line
    'linewidth': 1,         # width
    'linestyle': '-'        # style
}

# Median setting for children
median_style1 = {
    'color': 'darkslateblue',   # color of the line
    'linewidth': 1,         # width
    'linestyle': '-'        # style
}

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21, 7))


axes[0].bar(groups, means, color=['cadetblue', 'rosybrown'])
axes[1].boxplot([charges_lower_bmi, charges_normal_bmi, charges_higher_bmi, charges_obesity_bmi], patch_artist=True,
                boxprops=dict(facecolor='lightsteelblue'), medianprops=median_style)
axes[2].boxplot([charges_no_ch, charges_onetwo_ch, charges_threemore_ch], patch_artist=True,
                boxprops=dict(facecolor='thistle'), medianprops=median_style1)


# axes for chart with smokers
axes[0].set_title('Average Charges by Smoking Status', fontsize=14, fontweight='bold', pad = 15)
axes[0].set_xlabel('Smoking Status', fontsize=12, labelpad = 7)
axes[0].set_ylabel('Average Medical Charges ($)', fontsize=12, labelpad = 7)
axes[0].tick_params(axis='x', rotation=0)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# axes for chart with bmi
axes[1].set_title('Charges Distribution by BMI Category', fontsize=14, fontweight='bold', pad = 15)
axes[1].set_xlabel('BMI Category', fontsize=12, labelpad = 7)
axes[1].set_ylabel('Medical Charges ($)', fontsize=12, labelpad = 7)
axes[1].set_xticklabels(['Lower BMI', 'Normal BMI', 'Higher BMI', 'Obesity'])
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

# axes for chart with childrens number
axes[2].set_title('Charges Distribution by Number of Children', fontsize=14, fontweight='bold', pad = 15)
axes[2].set_xlabel('Number of Children', fontsize=12, labelpad = 7)
axes[2].set_ylabel('Medical Charges ($)', fontsize=12, labelpad = 7)
axes[2].set_xticklabels(['No kids', '1-2 kids', '3 and more kids'])
axes[2].grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

---
**We need to answer the question: is the difference in spending between different groups statistically significant?**
***

```python
# t-test for 'smoker', because we have only two groups

non_smoker = insurance[insurance['smoker_encoded'] == 0]['charges']
smoker = insurance[insurance['smoker_encoded'] == 1]['charges']

t_statistic, p_value = stats.ttest_ind(non_smoker, smoker)
print(f'''T-statistic: {t_statistic:.2f}
P-value: {p_value:.4f}''')
```

---
**Conclusions for t-Test for smokers/non-smokers**:
- The difference in average costs is statistically significant.
- The difference is significant and smoking status should be a key factor in pricing.
---

```python
# ANOVA for 'bmi' and 'children', because we have 3 and more groups

# ANOVA for 'bmi'
f_value, p_value = stats.f_oneway(charges_lower_bmi, charges_normal_bmi, charges_higher_bmi, charges_obesity_bmi)
print(f'''ANOVA for BMI:
F-value: {f_value:.2f}
P-value: {p_value:.4f}''')
```

---
**Conclusions for ANOVA for BMI level**:<br>
BMI category is an important factor to consider when calculating insurance premiums, because the difference in average medical costs between BMI categories is statistically significant
***

```python
# ANOVA for 'number of children'
f_value, p_value = stats.f_oneway(charges_no_ch, charges_onetwo_ch, charges_threemore_ch)
print(f'''ANOVA for number of children:
F-value: {f_value:.2f}
P-value: {p_value:.4f}''')
```

---
**Conclusions for ANOVA for Number of Children**:
- The difference in average medical expenses between groups with different numbers of children is statistically significant.

- At least one of the "number of children" groups has average expenses that are significantly different from the average expenses of the other groups.
---

```python
# ANOVA for 'region'
f_value, p_value = stats.f_oneway(insurance['region_northeast'], insurance['region_northwest'], insurance['region_southeast'], insurance['region_southwest'])
print(f'''ANOVA for region:
F-value: {f_value:.2f}
P-value: {p_value:.4f}''')
```

---
****Conclusions for ANOVA for Regions**:
- The difference you observe in average spending between the four regions is not statistically significant.

- Differences in average spending between regions are likely due to random sampling or random fluctuations, rather than fundamental differences within the regions themselves.
---

```python
# Encode smoker as binary
insurance['smoker_binary'] = insurance['smoker'].map({'yes':1, 'no':0})

# Define features and target
X = insurance[['bmi', 'smoker_binary', 'children']]
X = sm.add_constant(X)  # Adds intercept term
y = insurance['charges']

# Fit regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Predict charges using the fitted model
insurance['predicted_charges'] = model.predict(X)

plt.figure(figsize=(8,6))
plt.scatter(insurance['charges'], insurance['predicted_charges'], alpha=0.5)
plt.plot([insurance['charges'].min(), insurance['charges'].max()],
         [insurance['charges'].min(), insurance['charges'].max()],
         color='red', linestyle='--')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Actual vs Predicted Medical Charges')
plt.grid(True)
plt.show()
```
